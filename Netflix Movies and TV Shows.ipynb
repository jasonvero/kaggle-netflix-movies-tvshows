## Task 2: What year did Netflix release the most content to its platform?

```python
netflix_data.head()
```

```python
# info()

netflix_data.info()
```

```python
# pd.to_datetime() - converting an object column with a date value, to a date format

netflix_data['date_time'] = pd.to_datetime(netflix_data['date_added'])

netflix_data['date_time'].head()
```

```python
netflix_data.info()
```

```python
# dt

netflix_data['date_added_year'] = netflix_data['date_time'].dt.year
```

```python
netflix_data.head(2)
```

```python
netflix_data['count'] = 1

netflix_yearly_content_summed = netflix_data.groupby(['date_added_year']).sum().reset_index()[['date_added_year', 'count']]
```

```python
netflix_yearly_content_summed.sort_values(['count'], ascending = False)
```

#### Task 2 Answer: Year 2016, Netflix released 2016 mediums of content

## Task 4: What is the movie with the longest title in the data set?

```python
netflix_data.head()
```

```python
# movies_data = netflix_data[netflix_data['type'] == 'Movie'] -- previous code ran in earlier cell to isolate for only 'Movie' types

movies_data.head()
```

```python
# len() -- function is used to compute the length of each element in the Series/Index

movies_data['title_length'] = len(movies_data['title']) # -- this will count the number of rows, which is not what we're looking for 
```

```python
# drop(columns = ['']) -- drop a column from a df

movies_data.drop(columns = ['title_length'])
```

```python
movies_data.head()
```

```python
movies_data['title_length'] = [len(title) for title in movies_data['title']]
```

```python
movies_data.head()
```

```python
top_title_length = movies_data.sort_values(['title_length'], ascending = False)
```

```python
top_title_length
```

```python
# iloc -- Purely integer-location based indexing for selection by position.
top_title_length.iloc[0]['title']
```

## Task 5: Which actor/actress shows up most frequently in the Netflix data set

```python
netflix_data[['title', 'cast']]
```

```python
# for loop to iterate through the data frame and count things into a dictionary

title_and_cast = netflix_data[['title', 'cast']]
```

```python
title_and_cast = title_and_cast.dropna()
```

```python
title_and_cast.head()
```

```python
# sanity check to make sure your for loop is returning what's expected
for index, row in title_and_cast.iterrows():
    print(index)
    print()
    print(row)
    break
```

```python
# itterrows()
# split()
# strip()
# lower()

cast_names_counter = {} # dictionary

for index, row in title_and_cast.iterrows():
    movie_cast = row['cast']
    movie_cast_split = movie_cast.split(',')
    movie_cast_stripped = [name.strip().lower() for name in movie_cast_split]
    print(movie_cast_stripped)
    print()
    if index == 5:
        break
```

```python
# sorted()
# items()
# lambda
# dict()

cast_names_counter = {} # dictionary

for index, row in title_and_cast.iterrows():
    movie_cast = row['cast']
    movie_cast_split = movie_cast.split(',')
    movie_cast_stripped = [name.strip().lower() for name in movie_cast_split]
    for name in movie_cast_stripped:
        cast_names_counter[name] = cast_names_counter.get(name, 0) + 1
        
dict(sorted(cast_names_counter.items(), key = lambda item: item[1], reverse = True))
```

#### Task 5 Answer: Anupam Kher
Appeared in 43 movies
