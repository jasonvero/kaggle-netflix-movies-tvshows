## Task 1: What is the most popular release year for movies on Netflix

```python
# My attempt 

netflix_data['release_year'].value_counts()
```

```python
# Need to filter for just movies, not movies and tv shows

movies_data = netflix_data[netflix_data['type'] == 'Movie']
```

```python
# Sanity check that this worked

# unique() -- Return unique values based on a hash table

movies_data['type'].unique()
```

```python
# Sanity check that this worked

movies_data['type'].value_counts()
```

```python
movies_data
```

```python
# Tutorial way with 
# groupby()
# reset_index()

movies_data['count'] = 1

tut_release_years_summed = movies_data.groupby(['release_year']).sum().reset_index()[['release_year', 'count']]
```

```python
# My way with value_counts()
# Issue - not a df? Can't do a sort_values() since there are no column headers

ji_release_years_summed = movies_data['release_year'].value_counts()
ji_release_years_summed
```

```python
# sort_values()
# Answer: release_year 2017 or 2018 with 767 movie count each
tut_release_years_summed.sort_values(['count'], ascending = False)
```
